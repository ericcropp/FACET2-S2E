# Import and set up basic lattice

```python
from UTILITY_quickstart import *
```

```python
importedDefaultSettings = loadConfig("setLattice_configs/2024-10-22_oneBunch_baseline.yml")
```

```python
tao = initializeTao(
    #inputBeamFilePathSuffix = importedDefaultSettings["inputBeamFilePathSuffix"],
    inputBeamFilePathSuffix = '/beams/2024-12-11_Impact_OneBunch/2024-12-11_oneBunch.h5',
    
    csrTF = True,
    #numMacroParticles=1e5,
)

setLattice(tao, **importedDefaultSettings)
```

```python
trackBeam(tao, 
          trackStart = "L0AFEND", 
          #trackEnd = "L0BFEND", 
          trackEnd = "ENDBC14_2",
          verbose = True,
          **importedDefaultSettings
         )
```

```python
PInput = getBeamAtElement(tao, "L0BFEND")

display(plotMod(PInput, 'x', 'xp',  bins=300))
display(plotMod(PInput, 'x', 'energy',  bins=300))


print(PInput)
print(f"""Emittance: {emittance(PInput, plane = "x", fraction = 0.9)}, {emittance(PInput, plane = "y", fraction = 0.9)} m-rad""")
```
