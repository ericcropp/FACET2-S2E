```python
from UTILITY_quickstart import *
```

```python
importedDefaultSettings = loadConfig("setLattice_configs/defaults.yml")

tao = initializeTao(
#    inputBeamFilePathSuffix = inputBeamFilePathSuffix,
    
#    csrTF = csrTF,
#    numMacroParticles=1e5,
    scratchPath = "/tmp",
    randomizeFileNames = True
)
```

```python
setLattice(tao, **importedDefaultSettings)

launchTwissCorrection(tao)
```

```python
baselineTransport = getMatrix(tao, "BEGBC20", "MFFF", print = True)
```

```python
quadsToTweak = "Q1EkG"

for quadsToTweak in ["Q1EkG", "Q2EkG", "Q3EkG", "Q4EkG", "Q5EkG", "Q6EkG"]:
    print(quadsToTweak)
    setLattice(tao, **(importedDefaultSettings | { quadsToTweak : importedDefaultSettings[quadsToTweak] + 1 }))
    
    newMatrix = getMatrix(tao, "BEGBC20", "MFFF")
    deltaMatrix = newMatrix - baselineTransport
    displayMatrix(deltaMatrix)
           
```

```python
quadsToTweak = "Q1EkG"

for quadsToTweak in ["Q1EkG", "Q2EkG", "Q3EkG", "Q4EkG", "Q5EkG", "Q6EkG"]:
    print(quadsToTweak)
    setLattice(tao, **(importedDefaultSettings | { quadsToTweak : importedDefaultSettings[quadsToTweak] * 1.01 }))
    
    newMatrix = getMatrix(tao, "BEGBC20", "MFFF")
    deltaMatrix = newMatrix - baselineTransport
    displayMatrix(deltaMatrix)
```
