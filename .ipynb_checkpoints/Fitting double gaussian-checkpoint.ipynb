```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

# Step 1: Read the CSV file
file_path = '~/TEST_07605 - vertical projection data - no median subtraction.csv'
data = pd.read_csv(file_path)

# Step 2: Extract the first row of data
first_row = data.iloc[200]

# Step 3: Create a scatter plot of the first row
x = np.arange(len(first_row))
y = first_row.values

plt.scatter(x, y, label='Data')

# Step 4: Define a Gaussian function
def gaussian(x, amp, cen, wid):
    return amp * np.exp(-(x - cen)**2 / (2 * wid**2))

# Step 5: Fit a Gaussian to the data
popt, _ = curve_fit(gaussian, x, y, p0=[max(y), np.argmax(y), 1])

# Plot the Gaussian fit
x_fit = np.linspace(0, len(first_row) - 1, 1000)
y_fit = gaussian(x_fit, *popt)
plt.plot(x_fit, y_fit, label='Gaussian Fit', color='red')

# Adding labels and title
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Scatter Plot and Gaussian Fit of First Row')
plt.legend()

# Display the plot
plt.show()
```
